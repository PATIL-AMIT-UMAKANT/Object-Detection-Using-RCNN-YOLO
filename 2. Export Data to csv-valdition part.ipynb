```python
import pandas as pd 
import numpy as np
```

```python
df_images=pd.read_csv('val_images.csv')
df_images.head()
```

```python
df_categories=pd.read_csv('val_categories.csv')
df_categories.head()
```

```python
df_anno=pd.read_csv('val_annotations.csv')
df_anno.head()
```

```python
df_anno=df_anno[['image_id','bbox','category_id']]
df_anno
```

```python
df_images.columns
```

```python
df_images[['id','height']].values
```

```python
df_anno[['image_id']].values
```

```python
mapping = dict(df_images[['id','height']].values)
df_anno['height']=df_anno['image_id'].map(mapping)
df_anno
```

```python
mapping = dict(df_images[['id','height']].values)
df_anno['width']=df_anno['image_id'].map(mapping)
df_anno
```

```python
mapping = dict(df_images[['id','file_name']].values)
df_anno['file_name']=df_anno['image_id'].map(mapping)
df_anno
```

```python
df_categories.columns
```

```python
mapping = dict(df_categories[['id','name']].values)
df_anno['class']=df_anno['category_id'].map(mapping)
df_anno
```

```python
df_anno['bbox']=df_anno['bbox'].str.replace(r'\W',' ')
df_anno
```

```python
df_anno[['xmin','ymin','xmax','ymax']] = df_anno.bbox.str.split(expand=True)
df_anno
```

```python
df_anno=df_anno[['image_id','file_name','height','width','xmin','xmax','ymin','ymax','category_id','class']]
df_anno
```
